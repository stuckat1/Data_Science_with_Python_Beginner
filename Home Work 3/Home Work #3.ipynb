# Part I - Loading movie data

```python
import pandas as pd
import numpy as np

# load movie database
mnames = ['movieid', 'title', 'genres']
movies = pd.read_csv( ml_dir + r'\movies.dat', names = mnames, sep='::')

# load users database
unames  = [ 'userid', 'gender', 'age', 'occupation', 'zip']
users = pd.read_csv( ml_dir + r'\users.dat', names = unames, sep='::')

# load ratings database
rnames  = [ 'userid', 'movieid', 'rating', 'timestamp']
ratings = pd.read_csv( ml_dir + r'\ratings.dat', names = rnames, sep='::')
```

# VI - Construct a list of DataFrame

```python
def create_dict( db) :
    
    ret = {}
    
    for key, value in db.iteritems() :
        if key in master_keys :
            ret[key] = value
    
    return ret

master_keys = ['description', 'group', 'id', 'manufacturer']
db_list = [create_dict(uno) for uno in db]
meta_data = DataFrame(db_list)
```

```python
meta_data.head()
```

# VI - Create a nutrients DataFrame

```python
rez = []
for i, uno in enumerate(db) :
    dos = DataFrame( uno['nutrients'])
    dos['id'] = uno['id']
    rez.append( dos)

nutrients = pd.concat( rez, ignore_index = True)    
```

```python
nutrients.head()
```

# VI -  Drop duplicate entries in this DataFrame

```python
a = len(nutrients)
nutrients = nutrients.drop_duplicates()
b = len(nutrients)

print "Nutrients before %d, after %d" % (a,b)
```

# 1.4.4.3. Bar Plots

```python
%matplotlib inline
import pylab as pl
import numpy as np

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

pl.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
pl.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

# Do top graph
for x, y in zip(X, Y1):
    pl.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

# Do bottom graph
for x, y in zip(X, Y2):
    pl.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va='top')

# adjust ticks
pl.xlim(-.5, n)
pl.xticks(())
pl.ylim(-1.25, 1.25)
pl.yticks(())
```

# 1.4.4.4. Contour Plots

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

B = pl.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=pl.cm.hot)
C = pl.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
pl.clabel(C, inline=1, fontsize=10)

pl.xticks(())
pl.yticks(())
pl.show()
```

# 1.4.4.5. Imshow

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 10
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

pl.axes([0.025, 0.025, 0.95, 0.95])
pl.imshow(Z, interpolation='nearest', cmap='bone', origin='lower')
pl.colorbar(shrink=.92)

pl.xticks(())
pl.yticks(())
pl.show()
```

# 1.4.4.7 Quiver Plots

```python
import numpy as np

n = 8
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)

pl.axes([0.025, 0.025, 0.95, 0.95])
pl.quiver(X, Y, U, V, R, alpha=.5)
pl.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=.5)

pl.xlim(-1, n)
pl.xticks(())
pl.ylim(-1, n)
pl.yticks(())

pl.show()
```
